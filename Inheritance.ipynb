```python
# Multiple inheritance

class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def show_display(self):
        print("I am from Person class")
        print(f"{self.name} and {self.age}")

class Student(Person):  # ✅ Inherit from Person to use super().__init__()
    def __init__(self, name, age, branch):
        super().__init__(name, age)
        self.branch = branch

    def show_display(self):
        super().show_display()
        print(f"I am from Student class from {self.branch}")

class Employee(Person, Student):
    def __init__(self, name, age, branch, job, salary):
        Person.__init__(self, name, age)
        Student.__init__(self, name, age, branch)  # ✅ Initialize Student to set branch
        self.job = job
        self.salary = salary

    def show(self):
        super().show_display()
        print("I am from Employee class")

# Object creation
e = Employee("sr", 25, "ECE", "Software", 100000)
s = Student("Ravi", 22, "CSE")

# Output
print(e.branch)
e.show()
s.show_display()
```
