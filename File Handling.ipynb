```python
#file created writing
f=open("test.txt",'w+')
string="""
Welcome to our collection of sample PDF (Portable Document Format) files. 
These files are perfect for testing, development, 
and educational purposes. Whether you’re a developer, tester, or educator,
our diverse range of PDF samples will meet your needs.
"""

f.write(string)
f.seek(100)
print(f.read())
f.close()
```

```python
with open("test.txt",'r') as f:
    print(f.read(100))
    print('*'*20)
    print(f.readline())
    print(f.readlines())
```
