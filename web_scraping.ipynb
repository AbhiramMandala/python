```python
#pip install requests
#pip install bs4
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
search_ele="oppo"
flipkart_url=f"https://www.flipkart.com/search?q={search_ele}"
headers={"User-Agent": "Mozilla/5.0"}
webpage=requests.get(flipkart_url,headers=headers)
webpage
```

```python
print(webpage.text)
```

```python
web_res=bs(webpage.text,'html.parser')
web_res
```

```python
products=web_res.find_all("div",{'class':"cPHDOP col-12-12"})
len(products)
```

```python
product=products[3]
product
```

```python
product.div.div.div.a['href']
```

```python
"https://www.flipkart.com"+product.div.div.div.a['href']
```

```python
product_links=[]
for product in products:
    try:
        link="https://www.flipkart.com"+product.div.div.div.a['href']
        product_links.append(link)
    except:
        pass
print(product_links)
        
```

```python
product_link=product_links[0]
headers={"User-Agent": "Mozilla/5.0"}
product_page=requests.get(product_link,headers=headers)
product_page
```

```python
product_res=bs(product_page.text,'html.parser')
product_res
```

```python
comments=product_res.find_all("div",{"class":"RcXBOT"})
len(comments)
```
