```python
from abc import ABC,abstractmethod
class Shapes(ABC):
    @abstractmethod
    def perimeter(self):
        pass
    def area(self):
        pass
class Square(Shapes):
    def __init__(self, side):
        self.side = side
    def perimeter(self):
        return 4 * self.side
    def area(self):
        return self.side ** 2
class Rectangle(Shapes):
    def __init__(self, length, breadth):
        self.length = length
        self.breadth = breadth
    def perimeter(self):
        return 2 * (self.length + self.breadth)
    def area(self):
        return self.length * self.breadth
import math
class Circle(Shapes):
    def __init__(self, radius):
        self.radius = radius
    def perimeter(self):
        return 2 * math.pi * self.radius
    def area(self):
        return math.pi * self.radius ** 2
s = Square(5)
r = Rectangle(4, 6)
c = Circle(3)


print(f"Square  : Area is {s.area()} and Perimeter {s.perimeter()}")
print(f"Rectangle: Area is {r.area()} and Perimeter {r.perimeter()}")
print(f"Circle: Area is {c.area():.2f} and Perimeter {c.perimeter():.2f}")
```
