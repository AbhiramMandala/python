```python
import random 
print(random.random())#It generates random number between 0-1(1-excluded)
print(random.randint(1,6))#It generates rndom number between start and end value
l=[10,20,30,40,50,60,70,80,90]
print(random.choice(l))#It returns the random values among the seq
print(random.choices(l,k=3))#It returns the k-sized random seq among the seq
print(random.sample(l,k=3))#It returns the k-sized random values seq the seq
print(random.shuffle(l))
print(l)
```

```python
import math as m
print(m.ceil(7.001)) #it returns the upperbound value
print(m.floor(7.9001))  #it returns the lowerbound value
print(m.ceil(7.00000_00000_0000_1))
print(m.ceil(7.00000_00000_00000_1))
print(m.factorial(10))#It returns the factorial of a number
print(m.log10(100))   #it returns the log with 10 value
print(m.log2(4))
print(m.pi)
print(m.pow(2,10))
print(m.radians(45))
print(m.tan(m.radians(45)))
```
