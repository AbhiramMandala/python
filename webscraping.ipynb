```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd

search_ele = input()
flipkart_url = f"https://www.flipkart.com/search?q={search_ele}"
# table creation
columns =['Response Product','Username','Rating','Comment Header','Comment Body','Location','Commanted Ago']
df = pd.DataFrame([], columns = columns)
# get the content inside the url
headers = {"User-Agent": "Mozilla/5.0"}
response = requests.get(flipkart_url, headers = headers)
web_page = bs(response.text, 'html.parser')
products = web_page.find_all('div', {'class':'cPHDOP col-12-12'})
product_links=[]
for product in products:
    try:
        product_link = "https://www.flipkart.com"+product.div.div.div.a['href']
        product_links.append(product_link)
    except:
        pass
for product_link in product_links:
    product_page = requests.get(product_link,  headers = headers)
    product_response = bs(product_page.text, 'html.parser')
    comment_boxes = product_response.find_all('div', {'class':'RcXBOT'})
    for comment_box in comment_boxes:
        try:
            rating = comment_box.div.div.div.div.text
        except:
            rating = "No Rating"
        try:
            comment_header = comment_box.div.div.div.p.text
        except:
            comment_header = "No comment Header"
        try:
            comment_body = comment_box.div.div.find_all('div',{'class':'row'})[1].div.text
        except:
            comment_body = "No comment body"
        try:
            username = comment_box.div.div.find('div', {'class':'row gHqwa8'}).div.p.text
        except:
            username = "Username not found"
        try:
            location = comment_box.div.div.find('div', {'class':'row gHqwa8'}).div.find_all('p')[1].find_all('span')[1].text
        except:
            location = "No comment posted Location"
        try:
            time = comment_box.div.div.find('div', {'class':'row gHqwa8'}).div.find_all('p')[2].text
        except:
            time = "No Comment posted time"

        product_comment = {
            'Response Product': product_response.text[:50],
            'Username': username,
            'Rating': rating,
            'Comment Header': comment_header,
            'Comment Body': comment_body,
            'Location': location,
            'Commanted Ago': time
        }
        df.loc[len(df.index)] = list(product_comment.values())
# save as csv file
df.to_csv(f"{search_ele}.csv", header = True, index = False)
print("Table saved successfully")
```
